```python
def remove_html_tags(text):
    """Remove HTML tags from a string"""
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()

output['content'] = output['content'].apply(remove_html_tags)
```

## Tokenization:
Split text into individual words or tokens.

```python
nltk.download()
```

```python
df['tokens'] = df['content'].apply(lambda c: nltk.word_tokenize(c))
```

```python
df.tokens.sample().values
```

## Stop-word removal:
Remove common stop words (e.g., "the," "and," "in") that do not provide valuable information for analysis.

```python
stop_words_en = stopwords.words('english')
stop_words_fr = stopwords.words('french')
stop_words_ar = stopwords.words('arabic')
```

```python
df['tokens_without_sw'] = df['tokens'][df['language']=='en'].apply(lambda token_list: [word for word in token_list if word not in stop_words_en])
df['tokens_without_sw'][df['language']=='fr'] = df['tokens'][df['language']=='fr'].apply(lambda token_list: [word for word in token_list if word not in stop_words_fr])
df['tokens_without_sw'][df['language']=='ar'] = df['tokens'][df['language']=='ar'].apply(lambda token_list: [word for word in token_list if word not in stop_words_ar])
```

## Stemming or Lemmatization:
Reduce words to their root forms to handle variations (e.g., "running" to "run").

```python
stemmer_en = SnowballStemmer('english')
stemmer_fr = SnowballStemmer('french')
stemmer_ar = SnowballStemmer('arabic')
```

```python
df['stemmed_words'] = df['tokens_without_sw'][df['language']=='en'].apply(lambda x: [stemmer_en.stem(y) for y in x])
df['stemmed_words'][df['language']=='fr'] = df['tokens_without_sw'][df['language']=='fr'].apply(lambda x: [stemmer_fr.stem(y) for y in x])
df['stemmed_words'][df['language']=='ar'] = df['tokens_without_sw'][df['language']=='ar'].apply(lambda x: [stemmer_ar.stem(y) for y in x])
```
